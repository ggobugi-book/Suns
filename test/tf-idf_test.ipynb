```python
from collections import Counter
from konlpy.tag import Kkma
from konlpy.utils import pprint
from konlpy.tag import Twitter
from konlpy.tag import Komoran
from gensim.models import Word2Vec
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
```

```python
import re

file = open('./mujung.txt','r',encoding='utf-8')
lines = file.read()

sentences = re.split('(?<=[0-9]\n)|(?<=[0-9][0-9]\n)|(?<=[0-9][0-9][0-9]\n)', lines)

sent=[]
for stuff in sentences:
        sent.append(stuff)
```

```python
twitter = Twitter()
```

```python
sentences_tag = []
for sentence in sent:
    morph = twitter.pos(sentence)
    sentences_tag.append(morph)
```

```python
noun_adj_list = []
noun_adj=[]

for sentence1 in sentences_tag:
    for word, tag in sentence1:
        if tag in ['Noun','Adjective']:
            noun_adj.append(word)
```

```python
as_one = ''
for document in noun_adj:
    as_one = as_one + ' ' + document
words = as_one.split()
```

```python
counts = Counter(words)
counts
```

```python
vocab = sorted(counts, key=counts.get, reverse=True)
vocab
```

```python
import pandas as pd
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.manifold import TSNE
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
 
# this steps generates word counts for the words in your docs
word_count_vector=cv.fit_transform(noun_adj)
```

```python
vector = CountVectorizer()
print(vector.fit_transform(noun_adj).toarray()) # 코퍼스로부터 각 단어의 빈도 수를 기록한다.
print(vector.vocabulary_) # 각 단어의 인덱스가 어떻게 부여되었는지를 보여준다.
```

```python
tfidfv = TfidfVectorizer().fit(noun_adj)
print(tfidfv.transform(noun_adj).toarray())
print(tfidfv.vocabulary_)
```

```python
tfidf = TfidfVectorizer(max_features = 10, max_df=0.95, min_df=0)

#generate tf-idf term-document matrix
A_tfidf_sp = tfidf.fit_transform(noun_adj)
```

```python
tfidf_dict = tfidf.get_feature_names()
print(tfidf_dict)
```

```python
data_array = A_tfidf_sp.toarray()
data = pd.DataFrame(data_array, columns=tfidf_dict)
data.shape
```
