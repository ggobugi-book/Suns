```python
from collections import Counter
from konlpy.tag import Kkma
from konlpy.utils import pprint
from konlpy.tag import Twitter
from konlpy.tag import Komoran
from gensim.models import Word2Vec
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
```

```python
import re

#파일 읽어오기
file = open('./Thelittleprince.txt','r',encoding='utf-8')
line = file.read()

#장(chapter)마다 나눔
sentences = re.split('(?<=[2-9]\n)|(?<=[0-9][0-9]\n)|(?<=[0-9][0-9][0-9]\n)', line)

#Komoran 형태소 분석을 위한 전처리
sent=[]
for stuff in sentences:
#         stuff = stuff.replace("'","").replace("?","").replace(".","").replace("'","").replace('"'
# ,"").replace("!","").replace("-","").replace('"',"").replace("\n","")
        stuff = re.sub("\n","", line).strip()
        sent.append(stuff)
```

```python
print(sent[20])
```

```python
# twitter = Komoran(userdic='./user_dic.txt')
twitter = Twitter()
```

```python
#형태소분석
sentences_tag = []
for sentence in sent:
    morph = twitter.nouns(sentence)
    sentences_tag.append(morph)
```

```python
print(sentences_tag[0])
```

```python
#명사 추출
noun_adj=[]

#komoran
# for sentence1 in sentences_tag:
#     for word, tag in sentence1:
#         if tag in ['NNP']: 
#             noun_adj.append(word)

#twitter
for sentence1 in sentences_tag:
    for word, tag in sentence1:
        if tag in ['Noun']: #komoran
            noun_adj.append(word)
```
